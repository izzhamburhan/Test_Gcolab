```python
import pandas as pd
y = {'f(x)': [2,10,6,8], 'g(x)':[4,-4,5,6] }
x = [1,2,3,4]

graph = pd.DataFrame(y,x)
graph.plot(kind='line', grid=True, title='My Graph',ylabel='y',xlabel='x')
```
